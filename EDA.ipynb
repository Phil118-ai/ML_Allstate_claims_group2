```python
# ## 1. Import libraries and load data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load training dataset
df = pd.read_csv("data/train.csv")

# Quick overview
df.head()
```

```python
# ## 2. Inspect dataset structure

# Check number of rows and columns
print(df.shape)

# Column names (all variable names)
print(df.columns.tolist())

# Info on datatypes and non-null counts
df.info()
```

```python
# ## 3. Summary statistics

# Descriptive statistics for numeric variables
display(df.describe().T)

# For categorical variables
display(df.describe(include=['object', 'category']).T)
```

```python
# ## 4. Missing values

# Count missing values per column
print(df.isnull().sum().sort_values(ascending=False))

# Percentage of missing values
print((df.isnull().mean()*100).sort_values(ascending=False))
```

```python
# ## 10. Numeric distributions (hist + KDE, auto-chunked)
# Why: Skew, heavy tails, and clumping indicate need for transforms, binning, or robust models.

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cols = num_cols[:]
N = len(cols)
per_page = 12  # change if you want more/less per page

for i in range(0, N, per_page):
    subset = cols[i:i+per_page]
    rows = int(np.ceil(len(subset)/4))
    fig, axes = plt.subplots(rows, 4, figsize=(16, 3.5*rows))
    axes = np.ravel(axes)
    for ax, col in zip(axes, subset):
        sns.histplot(df[col].dropna(), bins=30, kde=True, ax=ax)
        ax.set_title(col)
    for ax in axes[len(subset):]:
        ax.remove()
    plt.tight_layout()
    plt.show()
```

```python
# ## 12. Outlier audit (IQR counts + boxplots for top offenders)
# Why: Extreme values can distort scaling, loss, and splits. Decide: cap/winsorize, transform, or leave for robust models.

def iqr_outlier_rate(s: pd.Series):
    s = s.dropna()
    if s.empty: return 0.0
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    if iqr == 0: return 0.0
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    return ((s < lower) | (s > upper)).mean()

outlier_rates = df[num_cols].apply(iqr_outlier_rate).sort_values(ascending=False)
display(outlier_rates.head(20).to_frame('outlier_rate'))

top_outliers = outlier_rates.head(8).index.tolist()
rows = int(np.ceil(len(top_outliers)/4)) if top_outliers else 0
if rows > 0:
    fig, axes = plt.subplots(rows, 4, figsize=(16, 3.5*rows))
    axes = np.ravel(axes)
    for ax, col in zip(axes, top_outliers):
        sns.boxplot(x=df[col], ax=ax)
        ax.set_title(f"{col} (~{outlier_rates[col]*100:.1f}% outliers)")
    for ax in axes[len(top_outliers):]:
        ax.remove()
    plt.tight_layout(); plt.show()
else:
    print("No numeric columns with notable IQR outliers.")
```

```python
# ## 13. Categorical distributions (top categories per feature)
# Why: High cardinality and long tails blow up OHE; compress rare levels before encoding.

cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

def plot_cat(col, top=20):
    vc = df[col].value_counts(dropna=False).head(top)
    plt.figure(figsize=(8, max(2, 0.4*len(vc))))
    sns.barplot(x=vc.values, y=vc.index, orient='h')
    plt.title(f"{col} (top {top})"); plt.xlabel('Count'); plt.ylabel('Category')
    plt.show()

for col in cat_cols[:20]:  # adjust how many to visualize
    plot_cat(col, top=20)
```

```python
# ## 14. High-cardinality categorical summary
# Why: Many unique labels → sparse matrices, leakage risk; consider hashing or target encoding for extreme cases.

if cat_cols:
    card = df[cat_cols].nunique(dropna=False).sort_values(ascending=False)
    display(card.to_frame('nunique').head(30))
else:
    print("No categorical columns detected.")
```

```python
# ## 15. Top correlated numeric pairs: quick pairplot (sampled)
# Why: Confirms redundancy visually; if relationships are near-linear, drop/combine or use regularization/PCA.

corr = df[num_cols].corr(numeric_only=True)
tri = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
pairs = (
    tri.abs()
    .stack()
    .sort_values(ascending=False)
    .head(5)  # top 5 strongest pairs
    .index.tolist()
)

pair_cols = sorted(list(set([p[0] for p in pairs] + [p[1] for p in pairs])))
if pair_cols:
    sample_df = df[pair_cols].dropna().sample(min(2000, df.shape[0]), random_state=42)
    sns.pairplot(sample_df, diag_kind="hist")
    plt.suptitle("Top correlated pairs (sampled)", y=1.02)
    plt.show()
else:
    print("No strong numeric pairs found to plot.")
```

```python
# ## 16. Duplicates overview
# Why: Exact duplicates inflate confidence, distort CV folds; remove if unjustified.

dup_rows = df.duplicated().sum()
print(f"Duplicate rows: {dup_rows} ({dup_rows/len(df)*100:.2f}%)")
```

```python
# ## 18. Numeric range/scale summary (for scaling decisions)
# Why: Huge scale differences suggest standardization/robust scaling before linear models or KNN/SVM.

rng = pd.DataFrame({
    'min': df[num_cols].min(),
    'p1': df[num_cols].quantile(0.01),
    'p50': df[num_cols].median(),
    'p99': df[num_cols].quantile(0.99),
    'max': df[num_cols].max(),
    'std': df[num_cols].std(),
}).round(3)

display(rng.sort_values('std', ascending=False).head(30))
```
